```python
df.drop(['id','ever_married','work_type','Residence_type'],axis=1,inplace=True)
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()

gen_encoder=lb.fit_transform(df['gender'])
df['gender']=gen_encoder
gen_encoder
```

```python
smoke_encoder=lb.fit_transform(df['smoking_status'])
df['smoking_status']=smoke_encoder
smoke_encoder
```

```python
df['gender']=df['gender'].astype(float)
df['smoking_status']=df['smoking_status'].astype(float)
```

```python
df.dtypes
```

```python
# missing value filling
df['bmi']=df['bmi'].fillna(df['bmi'].mean())
df.isna().sum()
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
sns.boxplot(x='avg_glucose_level',data=df)
```

```python
Q1=np.percentile(df['avg_glucose_level'],25,method='midpoint')
Q3=np.percentile(df['avg_glucose_level'],75,method='midpoint')
IQR=Q3-Q1
print(IQR)
```

```python
upper=Q3+1.5*IQR
upper_array=np.array(df['avg_glucose_level']>=upper)
print("Upper Bound:",upper)
print(upper_array.sum())

lower=Q1-1.5*IQR
lower_array=np.array(df['avg_glucose_level']<=lower)
print("Lower Bound:",lower)
print(lower_array.sum())
```

```python
sns.boxplot(x='bmi',data=df)
```

```python
Q1_bmi=np.percentile(df['bmi'],25,method='midpoint')
Q3_bmi=np.percentile(df['bmi'],75,method='midpoint')
IQR_bmi=Q3-Q1
print(IQR_bmi)
```

```python
upper2=Q3+1.5*IQR_bmi
upper_array2=np.array(df['bmi']>=upper2)
print("Upper Bound:",upper2)
print(upper_array2.sum())

#below Lower Bound
lower2=Q1_bmi-1.5*IQR_bmi
lower_array2=np.array(df['bmi']<=lower2)
print("Lower Bound:",lower2)
print(lower_array2.sum())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=50)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_test,y_pred)
score
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
cmd=ConfusionMatrixDisplay(confusion_matrix=cm)
cmd.plot()
#
```
